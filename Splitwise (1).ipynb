```python
import matplotlib.pyplot as plt
```

```python
def collect_transaction_data():
    # Step 1: Gather names
    n = int(input("How many people are there in this transaction? "))
    names = []
    for i in range(n):
        name = input(f"Enter name {i+1}: ")
        names.append(name)
    return names
```

```python
def plot_pie_chart(names, amounts, transaction_num):
    plt.figure(figsize=(8, 8))
    plt.pie(amounts, labels=names, autopct='%1.1f%%', startangle=140)
    plt.title(f'Money Distribution in Transaction {transaction_num}')
    plt.axis('equal')
    plt.show()
```

```python
# Collect data for multiple transactions
num_transactions = int(input("How many transactions would you like to enter? "))

for t in range(num_transactions):
    print(f"\nTransaction {t+1}")
    names = collect_transaction_data()
    total_money = float(input("Enter the total amount of money for this transaction: "))
    split_type = input("Type 'equal' for an equal split or 'unequal' for a specific amount split: ").lower()

    if split_type == 'equal':
        # Equal split
        equal_amounts = [total_money / len(names) for _ in names]
        print(f"\nEqual Split for Transaction {t+1}:", equal_amounts)
        plot_pie_chart(names, equal_amounts, t+1)

    elif split_type == 'unequal':
        # Unequal split
        print("\nEnter the specific amount for each person:")
        amounts = []
        for name in names:
            amount = float(input(f"Enter the amount for {name}: "))
            amounts.append(amount)

        total_specific_amounts = sum(amounts)
        if total_specific_amounts > total_money:
            print("Error: The total of specific amounts exceeds the total money available.")
        elif total_specific_amounts < total_money:
            print("Error: The total of specific amounts is less than the total money available.")
        else:
            print(f"\nUnequal Split for Transaction {t+1}:")
            plot_pie_chart(names, amounts, t+1)

    else:
        print("Invalid split type entered. Please enter either 'equal' or 'unequal'.")
```
